# Descriptive Statistic

Descriptive statistics help summarize and organize data so it becomes more understandable.

## Types of Descriptive Statistics
There are three categories for standard classification of descriptive statistics methods, each serving different purposes in summarizing and describing data. They help us understand:

- Where the data centers (**Measures of Central Tendency**)
- How spread out the data is (**Measure of Variability**)
- How the data is distributed (**Measures of Frequency Distribution**)

# Measures of Central Tendency

Central tendency refers to a single value that best represents a set of data, indicating the average or typical value within that data set. It's essentially a measure that helps describe the center of a distribution of data. The most common measures of central tendency are the **mean, median, and mode**.

The different measures of central tendency can be classified into three categories

# Objective and Functions of Averages

## Objective of Averages
The main objective of using averages is to represent a large set of data by a single value that gives a general idea or central tendency of the entire dataset. It simplifies complex data and makes interpretation easier.

## Functions of Averages:
- **Simplifies Data:**

    - Averages condense a large amount of data into a single representative figure.

    - Helps in summarizing large datasets clearly and concisely.

- **Comparison:**

    - Allows comparison between different groups, time periods, or variables.

    - Example: Comparing average income across two cities or average marks of students in two classes.

- **Basis for Decision-Making:**

    - Businesses, governments, and researchers use averages to make informed decisions.

    - Example: A company may increase production if the average demand is high.

- **Helps in Forecasting:**

    - Averages can be used to predict future trends.

    - Example: Using average sales data to forecast next month’s revenue.

- **Measures Central Tendency:**

    - Indicates the "central" or "typical" value in a dataset.

    - Common measures: Mean, Median, and Mode.

- **Identifies Abnormalities or Outliers:**

    - By comparing individual values with the average, unusual or extreme data points can be identified.

- **Supports Statistical Analysis:**

    - Averages are foundational for many statistical techniques like standard deviation, correlation, regression, etc.

---
 
## Requisites of a Good Measure of Central Tendency

1. **Easy to Understand and Calculate**  
   - Should be simple and not complicated to compute or explain.

2. **Clearly Defined**  
   - Must have a precise definition and give consistent results when applied.

3. **Based on All Observations**  
   - Should use all data points in the dataset for accuracy.

4. **Unaffected by Extreme Values (Outliers)**  
   - Should not be distorted by very high or very low values (especially important for skewed data).

5. **Mathematically Treatable**  
   - Should allow for further mathematical and statistical calculations.

6. **Representative of the Data**  
   - Must reflect the overall trend or typical value of the data set.

7. **Stable in Sampling**  
   - Should show little variation between different samples drawn from the same population.

8. **Suitable for Comparison**  
   - Should allow for meaningful comparison between different datasets or groups.


# Use Cases for Mean, Median, and Mode

## **Mean (Average)**

The **mean** is best used when you want a general average and your data doesn't have extreme outliers.

### Examples:
1. **Calculating average grades** for a class of students to determine overall performance.
2. **Determining the average monthly income** of employees in a company.
3. **Finding the average temperature** over a week to describe weather trends.

---

## **Median**

The **median** is useful when the data set has outliers or is skewed.

### Examples:
1. **Analyzing household income** in a country where a few very rich individuals could skew the mean.
2. **Finding the middle value of property prices** to better represent the market in a city.
3. **Identifying the typical wait time** in a queue system, especially when some waits are unusually long.

---

## **Mode**

The **mode** is useful for finding the most common or frequent item in a data set.

### Examples:
1. **Determining the most common shoe size** sold in a store.
2. **Finding the most frequent diagnosis** in a medical study.
3. **Analyzing customer preferences** by identifying the most frequently ordered dish in a restaurant.

# Measures of Dispersion 

Dispersion in statistics is a way to describe how spread out or scattered the data is around an average value. It helps to understand if the data points are close together or far apart.

Dispersion shows the variability or consistency in a set of data. There are different measures of dispersion like range, variance, and standard deviation

![](https://scielo.org.za/img/revistas/jsaimm/v123n9/02f01.jpg)

## Types of Measures of Dispersion
Measures of dispersion can be classified into the following two types :  

- Absolute Measure of Dispersion
- Relative Measure of Dispersion

These measures of dispersion can be further divided into various categories. They have various parameters and these parameters have the same unit.

| **Type**         | **Measure**                 | **Description**                                                                 |
|------------------|-----------------------------|---------------------------------------------------------------------------------|
| Absolute         | Range                       | Difference between the maximum and minimum values.                             |
| Absolute         | Interquartile Range (IQR)   | Spread of the middle 50% of the data (Q3 - Q1).                                |
| Absolute         | Variance                    | Average of the squared differences from the mean.                              |
| Absolute         | Standard Deviation          | Square root of variance; indicates average deviation from the mean.           |
| Absolute         | Mean Absolute Deviation     | Average of the absolute differences from the mean.                             |
| Relative         | Coefficient of Range        | Ratio of the range to the sum of maximum and minimum values.                  |
| Relative         | Coefficient of Variation    | Ratio of standard deviation to mean, usually expressed as a percentage.       |
| Relative         | Coefficient of Quartile Dev.| Ratio of half of IQR to the sum of Q1 and Q3.                                 |
| Relative         | Relative Mean Deviation     | Ratio of mean absolute deviation to the mean.                                 |

# Variance and Standard Deviation

## Standard Deviation
Standard Deviation measures the dispersion of data points which defines the extent to which data points in a dataset deviate from the mean, providing a clear sense of the variability or spread within the data.

$$\sigma = \sqrt{\frac{\sum_i^N(X_i - \bar{X})^2}{N}}$$

$$\sigma = \text{Standard Deviation}$$
$$X_i = \text{ith observation}$$
$$\bar{X} = \text{Mean}$$
$$N = \text{Number of Observations}$$

**The term inside the square root is the *Variance***

- Standard Deviation of a *Population* vs *Sample*

![](https://media.geeksforgeeks.org/wp-content/uploads/20230605181802/Standard-Deviation-Formula.png "Standard deviation Formula")

## Variance
Variance basically tells us how spread out a set of data is. If all the data points are the same, the variance is zero. Any non-zero variance is considered positive. Low variance means the data points are close to the average (or mean) and to each other. High variance means the data points are spread out from the average and from each other.

$$\text{Standard Deviation} = \sqrt{\text{Variance}}$$

$$Therefore$$

$$\text{Variance} = \text{Standard Deviation}^2$$

$$Variance(\sigma^2) = \frac{\sum_i^N(X_i - \bar{X})^2}{N}$$

$$\sigma = \text{Standard Deviation}$$
$$X_i = \text{ith observation}$$
$$\bar{X} = \text{Mean}$$
$$N = \text{Number of Observations}$$

## Difference between Variance and Standard Deviation

|Variance	                                                           |Deviation (Standard Deviation)
|----------------------------------------------------------------------|---------------------------------------------------
|Measure of spread in a dataset.	                                   |Measure of average distance from the mean.
|Average of squared differences from the mean.	                       |Square root of the variance.
|σ^2 (sigma squared)	                                               |σ (sigma)
|Indicates the average squared deviation of data points from the mean. |Indicates the average distance of data points from the mean.


# Correlation Theory
A statistical tool that helps in the study of the relationship between two variables is known as Correlation.

Two Variables are said to be Correlated if a change in one causes a corresponding change in the other variable. 

# Correlation and Causation
If two variables are correlated, it could be for any of the following reasons:
 - Third-Party Influence
 - Mutual Dependence (Cause and Effect)
 - Pure Chance

## Significance of Correlation

 - It helps determine the degree of correlation between the two variables in a single figure.
 - It makes understanding of economic behavior easier and identifies critical variables that are significant. 
 - When two variables are correlated, the value of one variable can be estimated using the value of the other. This is performed with the regression coefficients.
 - In the business world, correlation helps in taking decisions. The correlation helps in making predictions which helps in reducing uncertainty. It is so because the predictions based on correlation are probably reliable and close to reality.

### 1) Scatter Diagram:
A simple and attractive method of measuring correlation by diagrammatically representing bivariate distribution for determination of the nature of the correlation between the variables is known as the Scatter Diagram Method.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def create_plot(x: list, y: list, corr_type: str):
    """Function to generate a scatter plot, best fit line, and correlation matrix"""
    # Plot scatter plot
    plt.scatter(x, y)

    # Trend line
    m, c = np.polyfit(x, y, 1)
    trend_y = [m * x_i + c for x_i in x]

    # Plot trend line
    plt.plot(x, trend_y, color='red', linestyle="--", label="Trend Line")

    # Labels
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(f"x vs y ({corr_type})")
    plt.legend()
    
    # Show plot
    plt.show()
    
    # Create DataFrame and display correlation matrix
    data = pd.DataFrame({'x': x, 'y': y})
    corr_matrix = data.corr()
    print("\nCorrelation Matrix:")
    print(corr_matrix)
```

#### Perfect Positive Correlation(+1):
As one variable increases, the other variable increases in exact proportion. The points lie exactly on a straight upward-sloping line.

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]         

create_plot(x, y, "Perfect Positive Correlation")
```

#### Perfect Negative Correlation(-1)

As one variable increases, the other variable decreases in exact proportion. The points lie exactly on a straight downward-sloping line.

```python
x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]

create_plot(x, y, "Perfect Negative Correlation")
```

#### Strong Positive Correlation(close to +1, but not perfect)
As one variable increases, the other variable tends to increase, but not in a perfectly straight line.

```python
x = [1, 2, 3, 4, 5]
y = [2, 5, 7, 8, 11]

create_plot(x, y, "Strong Positive Correlation")
```

#### Strong Negative Correlation(close to -1, but not perfect)
As one variable increases, the other tends to decrease, but not perfectly linearly

```python
x = [1, 2, 3, 4, 5]
y = [9, 7, 5, 3, 2]

create_plot(x, y, "Strong Negative Correlation")
```

#### No Correlation (correlation = 0)
There is no predictable relationship between the two variables. Points are scattered randomly.

```python
x = [1, 2, 3, 4, 5]
y = [7, 3, 8, 2, 10]
create_plot(x, y, "No Correlation")
```

#### Merits

- **Simplicity**: Scatter Diagram is a simple and non-mathematical method to study correlation between two variables.
- **First Step**: It is the first step of investigating the relationship between two variables.
- **Easily Understandable**: One can easily understand and interpret scatter diagrams. Besides, only at a single glance at the diagram, one can easily tell the presence or absence of correlation.
- **Not Affected by Extreme Items**: The size of extreme values does not affect the scatter diagram. It is a quality which is not present in most mathematical methods.

#### Demerits
- **Rough Measure**: Scatter diagram only gives a rough idea of the degree and nature of correlation between the given two variables. Therefore, it is only a qualitative expression rather than a quantitative expression.
- **Non-mathematical Method**: Like other methods of correlation, Scatter Diagram Method does not indicate the exact numerical value of correlation. 
- **Unsuitable for Large Observations**: If there are more than two variables, it becomes difficult to draw a scatter diagram.

### 2) Karl Pearson's Coefficient of Correlation
Karl Pearson's Coefficient of Correlation is also known as **Product Moment Correlation** or **Simple Correlation Coefficient**. This method of measuring the coefficient of correlation is the most popular and is widely used. It is denoted by **'r'**, where r is a pure number which means that r has no unit. 

*Coefficient of Correlation is calculated by dividing the sum of products of deviations from their respective means by their number of pairs and their standard deviations.*

**Should only be applied to contiguous variables**

$$ r = \frac{\text{Sum of Products of Deviations from their respective means}} {\text{Number of Pairs} \times \text{Standard Deviations of both Series}} $$

$$ r = \frac {\sum (xy)} {N \cdot \sigma_X \cdot \sigma_Y} $$

$$\text{N} = \text{Number of Pair of Observations}$$
$$\text{x} = \text{Deviation of X series from Mean}{(X - \bar{X})}$$
$$\text{y} = \text{Deviation of Y series from Mean}{(Y - \bar{Y})}$$
$$\sigma_X = \text{Standard Deviation of X series}{\sqrt{\frac{\sum x^2}{N}}}$$
$$\sigma_Y = \text{Standard Deviation of Y series}{\sqrt{\frac{\sum y^2}{N}}}$$
$$\text{r} = \text{Coefficient of Correlation}$$

#### Features

- **Knowledge of Direction of Correlation**: This method of measuring coefficient of correlation gives us knowledge about the direction of the relationship between two variables. In other words, it tells us whether the relationship between two variables is positive or negative.

- **Size of Correlation**: Karl Pearson’s Coefficient of Correlation indicates the size of the relationship between two variables. Besides, Correlation Coefficient ranges between -1 and +1.

- **Indicates Magnitude and Direction**: This method not only specifies the magnitude of the correlation between two variables but also specifies its direction. It means that, if two variables are directly related, then the correlation coefficient between the variables will be a positive value. However, if two variables are inversely related, then the correlation coefficient between the variables will be a negative value.

- **Ideal Measure**: As this method is based on the most essential statistical measure, such as standard deviation and mean, it is an ideal/appropriate measure.


*Note: The value of the Correlation Coefficient should always lie between -1 and +1.*

- When `r = +1`, *it means that there is perfect positive correlation.*
- When `r = -1`, *it means that there is perfect negative correlation.*
- When `r = 0`, *it means that there is no or zero correlation.*


#### Assumptions
The assumptions on which Karl Pearson’s Coefficient of Correlation is based are as follows:

- **Linear Relationship**: The first assumption in this method is that there is a linear relationship between the given two variables. It means that if the paired observations of the variables (say X and Y) are plotted on a scatter diagram, then it will always form a straight line.

- **Casual Relationship**: Another assumption is that there is no cause-and-effect relationship between the given two variables. However, the cause-and-effect relationship exists between the forces affecting these variables. Besides, if there is no such type of relationship between the variables, then the correlation is meaningless.

- **Normal Distribution**: It is also assumed that, if there are a large number of independent causes of some nature producing normal distribution, then the two given variables are affected by them. For example, Variables like demand, supply, height, weight, etc., are affected by multiple forces.

- **Error of Measurement**: If the error of measurement is reduced to the minimum, then the coefficient of correlation is more reliable.


#### Properties
- **Coefficient of Correlation is Independent of change of origin and scale of measurements**: Coefficient of Correlation is not affected by the change of origin and scale of measurement.

- **Coefficient of Correlation lies between -1 and +1**:  The property of r also serves as a useful check on the correctness of the calculations. If the value of r lies outside the range, then it would mean that there is some error in the calculations.

- **Zero Correlation**: If two variables (say X and Y) are independent of each other, in that case, the coefficient of correlation between them will be zero.

- **Measure of Linear Relationship**: The coefficient of correlation is a measure that helps in determining the linear relationship between two variables. If both the variables (say X and Y) increase or decrease together, then r will be positive. However, if one variable increases when the other variable decreases or vice-versa, then r will be negative.


#### Merits

- **Popular Method**: Karl Pearson’s Coefficient of Correlation is the most popular and widely used mathematical method to study the correlation between two variables.

- **Degree and Direction of Correlation**: The value of correlation coefficient not only summarises the degree of correlation but also its direction.

#### Demerits

- **Affected by Extreme Values**: If the values of the two variables are extreme, then it would have a great impact on the value of correlation coefficient.

- **Assumption of Linear Relationship**: While determining correlation coefficient, it is always assumed that there is a linear relationship without thinking whether the assumption is correct or not.

- **Possibility of Wrong Interpretation**:  While interpreting the value of coefficient of correlation using this method, one has to be very careful. It is because the chances of misinterpreting the coefficient are more. 

### 3) Spearman's Rank Correlation Coefficient
Spearman's Rank Correlation Coefficient or Spearman's Rank Difference Method or Formula is a method of calculating the correlation coefficient of qualitative variables and was developed in 1904 by Charles Edward Spearman. 

In other words, the formula determines the correlation coefficient of variables like beauty, ability, honesty, etc., whose quantitative measurement is not possible. Therefore, these attributes are ranked or put in the order of their preference. 

$$
\rho = 1 - \frac{6 \sum D^2}{N(N^2 - 1)}
$$

$$\rho = \text{Coefficient of rank correlation}$$
$$\text{D} = \text{Rank differences}$$
$$\text{N} = \text{Number of variables}$$

#### Properties
- The value of Spearman's Rank Correlation Coefficient (𝜌) lies between -1 and +1, inclusive.

- `𝜌 = + 1` *indicates a perfect positive monotonic relationship.*
- `𝜌 = - 1` *indicates a perfect negative monotonic relationship.*
- `𝜌 = 0` *indicates no monotonic relationship.*

**Monotonicity**:

Spearman's Rank Correlation measures the strength and direction of a monotonic relationship, which means that the variables do not need to have a linear relationship. They simply need to increase or decrease together in a consistent (monotonic) manner.
